```python
# Cell 3 — Extract toàn bộ .mbox → .eml (Colab-ready)

import mailbox, os, hashlib
from pathlib import Path

RAW = "/content/drive/MyDrive/phish_project/raw"
EML_DIR = os.path.join(RAW, "eml")
os.makedirs(EML_DIR, exist_ok=True)

mbox_files = [f for f in os.listdir(RAW) if f.endswith(".mbox")]

print("Tìm thấy .mbox để extract:", mbox_files)

def extract_mbox(mbox_path, out_dir):
    mbox = mailbox.mbox(mbox_path)
    count = 0
    print(f"Đang extract: {mbox_path} ...")
    for i, msg in enumerate(mbox):
        try:
            raw = msg.as_bytes()
            msg_id = msg.get('Message-ID') or msg.get('message-id') or f"{i}"
            hash_id = hashlib.md5((str(msg_id) + str(i)).encode()).hexdigest()
            out_file = os.path.join(out_dir, f"{Path(mbox_path).stem}_{hash_id}.eml")
            with open(out_file, "wb") as f:
                f.write(raw)
            count += 1
        except Exception as e:
            print("Lỗi extract:", e)
    print(f" → Hoàn tất {count} email từ file: {mbox_path}")

for mbox_name in mbox_files:
    mbox_path = os.path.join(RAW, mbox_name)
    extract_mbox(mbox_path, EML_DIR)

print("\nHoàn tất extract .mbox → .eml")
print("Số lượng file .eml:", len(os.listdir(EML_DIR)))
```

```python
# Cell 4 — FULL PARSER: subject + header + HTML + URL + attachments
# Output: processed/emails_unified.parquet

import os, re, json, hashlib
from pathlib import Path
import pandas as pd
import email
from email import policy
from bs4 import BeautifulSoup

BASE = '/content/drive/MyDrive/phish_project'
RAW = os.path.join(BASE, 'raw')
EML_DIR = os.path.join(RAW, 'eml')
PROC_DIR = os.path.join(BASE, 'processed')
os.makedirs(PROC_DIR, exist_ok=True)

print("PARSE FROM:", EML_DIR)

# Regex tìm URL
RE_URL = re.compile(r'(https?://[^\s"\'>]+)', re.I)

def html_to_text(html):
    if not html:
        return ""
    soup = BeautifulSoup(html, "lxml")
    for s in soup(["script", "style", "noscript"]):
        s.decompose()
    text = soup.get_text(" ")
    return re.sub(r"\s+", " ", text).strip()

def extract_urls(text):
    if not text:
        return []
    urls = RE_URL.findall(text)
    return list(dict.fromkeys(urls))  # unique + keep order

def parse_eml(path):
    """
    Full parser:
    - full headers
    - extract HTML + TEXT
    - extract URLs
    - extract attachments
    """
    try:
        raw = Path(path).read_bytes()
        try:
            msg = email.message_from_bytes(raw, policy=policy.default)
        except:
            msg = email.message_from_bytes(raw)

        hdr = dict(msg.items())

        subject = msg.get('subject', '') or ''
        from_    = msg.get('from', '') or ''
        to       = msg.get('to', '') or ''
        date     = msg.get('date', '') or ''
        msg_id   = msg.get('message-id') or msg.get('Message-ID') or ""

        body_text = ""
        body_html = ""
        attachments = []

        # Multipart emails
        if msg.is_multipart():
            for part in msg.walk():
                ctype = part.get_content_type()
                disp  = str(part.get_content_disposition())

                # Plain text
                if ctype == "text/plain" and disp != "attachment":
                    try:
                        body_text += part.get_content()
                    except:
                        body_text += part.get_payload(decode=True).decode(errors="ignore")

                # HTML
                elif ctype == "text/html" and disp != "attachment":
                    try:
                        body_html += part.get_content()
                    except:
                        body_html += part.get_payload(decode=True).decode(errors="ignore")

                # Attachments
                elif disp == "attachment" or part.get_filename():
                    attachments.append({
                        "filename": part.get_filename(),
                        "mime": ctype,
                    })

        else:
            # Singlepart messages
            ctype = msg.get_content_type()
            if ctype == "text/plain":
                body_text = msg.get_content()
            elif ctype == "text/html":
                body_html = msg.get_content()

        # Convert HTML → text
        html_text = html_to_text(body_html)

        # Extract URLs from BOTH text + html
        urls = extract_urls(body_text + "\n" + body_html)

        # Unique hash để dedupe
        raw_hash = hashlib.md5(raw).hexdigest()

        return {
            "message_id": msg_id,
            "raw_hash": raw_hash,
            "subject": subject,
            "from": from_,
            "to": to,
            "date": date,
            "headers": json.dumps(hdr),
            "body_text": body_text,
            "body_html": body_html,
            "body_html_text": html_text,
            "full_text": f"{subject} {body_text} {html_text}".strip(),
            "urls": json.dumps(urls),
            "attachments": json.dumps(attachments),
        }

    except Exception as e:
        print("Lỗi parse:", path, e)
        return None


# ======================
# PARSE TOÀN BỘ .eml
# ======================
eml_files = sorted([os.path.join(EML_DIR, f) for f in os.listdir(EML_DIR) if f.endswith(".eml")])
print("Tổng số file .eml sẽ parse:", len(eml_files))

rows = []
for i, p in enumerate(eml_files):
    if i % 200 == 0:
        print(f" → Đang parse: {i}/{len(eml_files)}")
    row = parse_eml(p)
    if row:
        rows.append(row)

df = pd.DataFrame(rows)
df = df.drop_duplicates(subset=["raw_hash"]).reset_index(drop=True)
print("Số email sau dedupe:", len(df))

# Lưu parquet
out_path = os.path.join(PROC_DIR, "emails_unified.parquet")
df.to_parquet(out_path, index=False)

print("\n✔ FULL PARSER HOÀN TẤT!")
print("Đã lưu file:", out_path)
print(df.head())
```

```python
# Cell 9 - Parse & chuẩn hoá Kaggle phishing_email.csv -> parquet chuẩn
import os, json, re, hashlib
import pandas as pd
from pathlib import Path
from tqdm.notebook import tqdm
RAW_KAGGLE = "/content/drive/MyDrive/phish_project/raw/kaggle"
IN_PATH = os.path.join(RAW_KAGGLE, "phishing_email.csv")   # <-- local path
OUT_DIR = "/content/drive/MyDrive/phish_project/processed"
os.makedirs(OUT_DIR, exist_ok=True)
print("Input file:", IN_PATH)

# 1) đọc nhanh, show thông tin
df = pd.read_csv(IN_PATH, low_memory=False)
print("Loaded kaggle file rows:", len(df))
print("Columns:", df.columns.tolist())
display(df.head())

# 2) map sang schema chuẩn
# Các tên cột phổ biến: ['subject','body_text','body_html','text','url_list','sender','label',...]
# Thử chọn cột phù hợp tự động
col_map = {}
for cand in ['subject','title','mail_subject']:
    if cand in df.columns:
        col_map['subject'] = cand
for cand in ['body_text','text','content','email_body','body']:
    if cand in df.columns and 'body_text' not in col_map:
        col_map['body_text'] = cand
for cand in ['body_html','html','html_body']:
    if cand in df.columns:
        col_map['body_html'] = cand
for cand in ['url_list','urls','links']:
    if cand in df.columns:
        col_map['urls'] = cand
for cand in ['sender','from','from_address']:
    if cand in df.columns:
        col_map['from'] = cand
# label
label_col = None
for cand in ['label','is_phish','class','target']:
    if cand in df.columns:
        label_col = cand
        break

print("Detected mapping:", col_map, "label_col:", label_col)

# 3) build canonical rows
def safe_get(r, k):
    return r[k] if (k in r and pd.notnull(r[k])) else ""

rows = []
URL_RE = re.compile(r'(https?://[^\s"\']+)', re.I)
print(" Starting row processing...")
for i, r in df.iterrows():
    subj = safe_get(r, col_map.get('subject','')) if col_map.get('subject') else ""
    body_text = safe_get(r, col_map.get('body_text','')) if col_map.get('body_text') else ""
    body_html = safe_get(r, col_map.get('body_html','')) if col_map.get('body_html') else ""
    # if urls column exists, try to parse it; else extract from text/html
    if 'urls' in col_map:
        raw_urls = safe_get(r, col_map['urls'])
        try:
            if isinstance(raw_urls, str) and (raw_urls.startswith('[') or raw_urls.startswith('"')):
                urls = json.loads(raw_urls)
            elif pd.isna(raw_urls) or raw_urls=="":
                urls = URL_RE.findall((body_text or "") + "\n" + (body_html or ""))
            else:
                # heuristics split by comma/space
                urls = re.findall(URL_RE, str(raw_urls))
        except Exception:
            urls = URL_RE.findall((body_text or "") + "\n" + (body_html or ""))
    else:
        urls = URL_RE.findall((body_text or "") + "\n" + (body_html or ""))
    full_text = ("{} {}".format(subj, body_text or body_html)).strip()
    label = None
    if label_col:
        lab = r[label_col]
        # try to normalize common labels: phishing/spam/ham
        if isinstance(lab, str):
            lab_l = lab.strip().lower()
            if lab_l in ['phish','phishing','1','true','yes']:
                label = 1
            elif lab_l in ['ham','legit','0','false','no','normal','notphish']:
                label = 0
            else:
                # numeric?
                try:
                    label = int(lab)
                except:
                    label = None
        else:
            try:
                label = int(lab)
            except:
                label = None

    raw_hash = hashlib.md5((full_text + str(r.get(col_map.get('from',''),""))).encode('utf-8', errors='ignore')).hexdigest()
    rows.append({
        "message_id": None,
        "raw_hash": raw_hash,
        "subject": subj,
        "from": safe_get(r, col_map.get('from','')),
        "to": "",
        "date": r.get('date',"") if 'date' in r else "",
        "headers": None,
        "body_text": body_text,
        "body_html": body_html,
        "body_html_text": "",   # keep empty (we'll fill later if needed)
        "full_text": full_text,
        "urls": json.dumps(urls),
        "attachments": json.dumps([]),
        "label": label,
        "source": "kaggle_phishing"
    })

# 4) to DataFrame and save
df_kag = pd.DataFrame(rows)
print("Parsed kaggle -> rows:", len(df_kag))
out_path = os.path.join(OUT_DIR, "kaggle_parsed.parquet")
df_kag.to_parquet(out_path, index=False)
print("Saved parsed kaggle to:", out_path)
display(df_kag.head())
```

```python
# CELL 12 — Parse SpamAssassin (easy_ham + spam) -> spamassassin_parsed.parquet
import os, hashlib, json, re
import pandas as pd
from pathlib import Path

RAW_SA = "/content/drive/MyDrive/phish_project/raw/spamassassin"
OUT_DIR = "/content/drive/MyDrive/phish_project/processed"
os.makedirs(OUT_DIR, exist_ok=True)
OUT = os.path.join(OUT_DIR, "spamassassin_parsed.parquet")

# regex để bắt URL
EMAIL_RE = re.compile(r'(https?://[^\s"\']+)', re.I)

def read_sa_folder(folder, label):
    rows = []
    p = Path(folder)
    if not p.exists():
        print("Folder not found, skip:", folder)
        return rows
    files = sorted([f for f in p.iterdir() if f.is_file()])
    for file in files:
        try:
            content = file.read_text(errors='ignore')
            urls = EMAIL_RE.findall(content)
            full_text = content.strip()
            raw_hash = hashlib.md5(full_text.encode('utf-8', errors='ignore')).hexdigest()
            rows.append({
                "message_id": None,
                "raw_hash": raw_hash,
                "subject": "",
                "from": "",
                "to": "",
                "date": "",
                "headers": None,
                "body_text": full_text,
                "body_html": "",
                "body_html_text": "",
                "full_text": full_text,
                "urls": json.dumps(urls),
                "attachments": json.dumps([]),
                "label": label,   # ham=0, spam=1
                "source": "spamassassin"
            })
        except Exception as e:
            # in lỗi nhưng không dừng toàn bộ quá trình
            print("Error parsing:", file, repr(e))
    return rows

# thư mục ham/spam thực tế trong kho bạn báo: 'easy_ham' và 'spam'
ham_dir = os.path.join(RAW_SA, "easy_ham")
spam_dir = os.path.join(RAW_SA, "spam")

rows = []
print("Checking folders:", ham_dir, spam_dir)
if os.path.exists(ham_dir):
    print("Parsing ham from:", ham_dir)
    rows += read_sa_folder(ham_dir, 0)
else:
    print("Ham folder missing:", ham_dir)

if os.path.exists(spam_dir):
    print("Parsing spam from:", spam_dir)
    rows += read_sa_folder(spam_dir, 1)
else:
    print("Spam folder missing:", spam_dir)

df_sa = pd.DataFrame(rows)
df_sa = df_sa.drop_duplicates(subset=["raw_hash"]).reset_index(drop=True)

# lưu kết quả
df_sa.to_parquet(OUT, index=False)
print("Parsed SpamAssassin saved:", OUT)
print("Số email spamassassin đã parse:", len(df_sa))
display(df_sa.head())
```

```python
#CELL 17 – TRAIN LinearSVC (FINAL PATCHED)

import os, json, re, hashlib, joblib
import numpy as np
import pandas as pd
from tqdm import tqdm
from scipy.sparse import hstack, csr_matrix

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC


# LOAD DATA

PROC = "/content/drive/MyDrive/phish_project/processed"
IN_FILE = f"{PROC}/emails_all_combined.parquet"
MODEL_DIR = "/content/drive/MyDrive/phish_project/models"
os.makedirs(MODEL_DIR, exist_ok=True)

print("Loading dataset:", IN_FILE)
df = pd.read_parquet(IN_FILE)
print("Total rows:", len(df))
print(df["label"].value_counts())

#  FIX MISSING LABELS + CLEAN TEXT

df["label"] = df["label"].fillna(0).astype(int)
df = df[df["label"].isin([0,1])]
df = df.dropna(subset=["full_text"])
df = df.reset_index(drop=True)

print("After clean:", df["label"].value_counts())

#  URL + HEADER FEATURE FUNCTIONS (FINAL PATCHED)

URL_RE = re.compile(r'(https?://[^\s"\']+)', re.I)

def compute_entropy(domain):
    if not domain: return 0.0
    from collections import Counter
    cnt = Counter(domain)
    probs = [v / len(domain) for v in cnt.values()]
    import math
    return -sum(p * math.log(p + 1e-12, 2) for p in probs)

def extract_url_features(urls_json):
    try:
        if isinstance(urls_json, str):
            urls = json.loads(urls_json)
            if not isinstance(urls, list):
                urls = []
        else:
            urls = []
    except:
        urls = []

    n_urls = len(urls)
    max_len = 0
    any_short = 0
    any_ip = 0
    entropies = []

    for u in urls:
        max_len = max(max_len, len(u))
        host = re.sub(r"^.*?//","",u).split("/")[0]

        if any(x in host for x in ["bit.ly","t.co","tinyurl","goo.gl","is.gd","ow.ly"]):
            any_short = 1

        if re.search(r"\b\d{1,3}(?:\.\d{1,3}){3}\b", host):
            any_ip = 1

        entropies.append(compute_entropy(host))

    mean_entropy = float(np.mean(entropies)) if entropies else 0.0

    return [n_urls, max_len, mean_entropy, any_short, any_ip]


# - PATCHED VERSION (KHÔNG BAO GIỜ LỖI)
def extract_header_features(hjson):
    # safe parse
    if hjson is None or hjson == "" or (isinstance(hjson, float) and np.isnan(hjson)):
        h = {}
    else:
        try:
            h = json.loads(hjson) if isinstance(hjson, str) else {}
        except:
            h = {}

    # spf/dkim
    text = json.dumps(h).lower()
    spf_fail  = 1 if "spf=fail" in text else 0
    dkim_fail = 1 if "dkim=fail" in text else 0

    # domain mismatch detection
    from_addr = h.get("From","") or h.get("from","") or ""
    rp        = h.get("Return-Path","") or h.get("return-path","") or ""

    def get_dom(addr):
        if not isinstance(addr, str): addr = str(addr)
        m = re.search(r"@([A-Za-z0-9.-]+)", addr)
        return m.group(1).lower() if m else ""

    d1 = get_dom(from_addr)
    d2 = get_dom(rp)

    # always safe boolean
    mismatch = 1 if (d1 and d2 and d1 != d2) else 0

    return [spf_fail, dkim_fail, mismatch]


#         BUILD NUMERIC FEATURE MATRIX (8 FEATURES)

print("Building numeric features...")

numeric_rows = []
for urls, hdr in tqdm(zip(df["urls"].fillna("[]"), df["headers"].fillna("{}")), total=len(df)):
    numeric_rows.append(
        extract_url_features(urls) + extract_header_features(hdr)
    )

numeric_arr = np.array(numeric_rows, dtype=float)
numeric_arr = np.nan_to_num(numeric_arr, nan=0.0)

scaler = StandardScaler()
numeric_scaled = scaler.fit_transform(numeric_arr)


#           TF-IDF VECTORIZATION (200k FEATURES)

print("Vectorizing TF-IDF...")

tfidf = TfidfVectorizer(
    max_features=200000,
    ngram_range=(1,2),
    strip_accents="unicode",
    lowercase=True,
    min_df=3,
    max_df=0.97
)

X_text = tfidf.fit_transform(df["full_text"].astype(str))
print("TF-IDF shape:", X_text.shape)

# combine all features
X = hstack([X_text, csr_matrix(numeric_scaled)], format="csr")
y = df["label"].values

print("Final feature shape:", X.shape)


#                  TRAIN / TEST SPLIT

print("Splitting 80/20...")
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)


#                   TRAIN LinearSVC

print("Training Linear SVC...")

svc = LinearSVC(class_weight="balanced", max_iter=10000)
params = {"C": [0.1, 0.5, 1, 3, 5]}

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

gs = GridSearchCV(svc, params, scoring="f1", cv=cv, n_jobs=-1, verbose=2)
gs.fit(X_train, y_train)

best = gs.best_estimator_

print("\n=== Best Params:", gs.best_params_)
print("=== Best CV F1:", gs.best_score_)


#                  EVALUATION

print("\n=== TEST SET REPORT ===")
y_pred = best.predict(X_test)
print(classification_report(y_test, y_pred, digits=4))

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(
    cm, annot=True, fmt="d", cmap="Blues",
    xticklabels=["Ham", "Phish"],
    yticklabels=["Ham", "Phish"]
)
plt.title("Confusion Matrix – LinearSVC (Hybrid Features)")
plt.show()

#               SAVE MODEL + ARTIFACTS

joblib.dump(best, f"{MODEL_DIR}/linear_svc_model.pkl")
joblib.dump(tfidf, f"{MODEL_DIR}/tfidf.pkl")
joblib.dump(scaler, f"{MODEL_DIR}/scaler.pkl")

metadata = {
    "numeric_feature_order": [
        "n_urls","max_url_len","mean_entropy","any_short","any_ip",
        "spf_fail","dkim_fail","header_mismatch"
    ],
    "tfidf_max_features": 200000,
    "model_type": "LinearSVC",
    "version": "1.0"
}
json.dump(metadata, open(f"{MODEL_DIR}/metadata.json","w"))

print("Saved model →", MODEL_DIR)

#             REALTIME PREDICT FUNCTION

def make_features(full_text, urls_list=None, headers_json="{}"):
    urls_json = json.dumps(urls_list) if isinstance(urls_list,(list,tuple)) else "[]"
    num = extract_url_features(urls_json) + extract_header_features(headers_json)
    num = scaler.transform([num])
    xt = tfidf.transform([full_text])
    return hstack([xt, csr_matrix(num)], format="csr")
def predict_email(full_text, urls=None, headers="{}"):
    Xs = make_features(full_text, urls, headers)
    label = best.predict(Xs)[0]
    return {"label": int(label)}
```

```python

# Cell 21 BUILD FEATURES (phải chạy trước Cell 1)


import os, json, joblib
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from scipy.sparse import hstack, csr_matrix

BASE = "/content/drive/MyDrive/phish_project"
PROC = os.path.join(BASE, "processed")

# Load dataset đã clean
df = pd.read_parquet(os.path.join(PROC, "emails_all_combined.parquet"))
print("Loaded:", len(df))

# ============ BUILD FULL_TEXT ============
df['full_text'] = (
    df['subject'].fillna("") + " " +
    df['body_text'].fillna("") + " " +
    df['body_html'].fillna("")
)

# ============ BUILD TF-IDF ================
print("Fitting TF-IDF...")
tfidf = TfidfVectorizer(
    max_features=150000,
    ngram_range=(1,2),
    stop_words="english"
)
X_text = tfidf.fit_transform(df['full_text'])

# ============ NUMERIC FEATURES =============
def compute_entropy(domain):
    from collections import Counter
    import math
    if not domain:
        return 0
    cnt = Counter(domain)
    probs = [v/len(domain) for v in cnt.values()]
    return -sum(p*math.log(p+1e-9,2) for p in probs)

import re, json as _json

def extract_features(row):
    try:
        urls = _json.loads(row['urls'])
        if not isinstance(urls, list): urls = []
    except:
        urls = []

    n_urls = len(urls)
    max_len = max([len(u) for u in urls], default=0)

    ent = []
    short = 0
    ip = 0

    for u in urls:
        host = re.sub(r"^.*?//","",u).split("/")[0]
        ent.append(compute_entropy(host))

        if any(x in u for x in ['bit.ly','tinyurl','t.co','ow.ly','goo.gl']):
            short = 1

        if re.search(r"\b\d{1,3}(\.\d{1,3}){3}\b", u):
            ip = 1

    mean_ent = float(np.mean(ent)) if ent else 0.0

    # headers
    try:
        h = _json.loads(row['headers'])
    except:
        h = {}

    hv = json.dumps(h).lower()
    spf = 1 if "spf=fail" in hv else 0
    dkim = 1 if "dkim=fail" in hv else 0

    from_addr = h.get("From","") or h.get("from","")
    ret_path = h.get("Return-Path","")

    def dom(a):
        m = re.search(r"@([A-Za-z0-9.-]+)", str(a))
        return m.group(1).lower() if m else ""

    mismatch = 1 if dom(from_addr) and dom(ret_path) and dom(from_addr)!=dom(ret_path) else 0

    return [n_urls, max_len, mean_ent, short, ip, spf, dkim, mismatch]

print("Building numeric features...")
num_features = df.apply(extract_features, axis=1, result_type="expand")
num_features.columns = ["n_urls","max_url_len","mean_entropy","any_short","any_ip","spf_fail","dkim_fail","header_mismatch"]

# Scale numeric
scaler = StandardScaler()
X_num = scaler.fit_transform(num_features.values)

# ============ COMBINE FEATURES =============
X_all = hstack([X_text, csr_matrix(X_num)], format="csr")
y = df['label'].astype(int).values

print("X_all shape:", X_all.shape)
print("y distribution:", np.bincount(y))

print(" DONE BUILDING FEATURES — giờ bạn có thể chạy CELL 1")
```

```python

# Cell 24 — Fine-tune DistilBERT (FIX FOR num_items_in_batch)


import os, numpy as np, pandas as pd, torch, random
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
    DataCollatorWithPadding,
    set_seed
)
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import precision_recall_fscore_support, accuracy_score

# Reference uploaded file (if needed): /mnt/data/untitled0.py

# Disable W&B to avoid API key prompt
os.environ["WANDB_DISABLED"] = "true"

# ---------- CONFIG ----------
BASE = "/content/drive/MyDrive/phish_project"
PROC = os.path.join(BASE, "processed")
MODEL_DIR = os.path.join(BASE, "models")
OUT_BERT = os.path.join(MODEL_DIR, "distilbert_stage2")
os.makedirs(OUT_BERT, exist_ok=True)

TRAIN_PATH = os.path.join(PROC, "stage2_train.parquet")
VAL_PATH   = os.path.join(PROC, "stage2_val.parquet")

SEED = 42
set_seed(SEED)
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

NUM_EPOCHS = 8
BATCH = 8
ACCUM_STEPS = 2
LR = 3e-5
MAX_LEN = 256

# ---------- LOAD DATA ----------
print("🔄 Loading Stage2 datasets...")
train_df = pd.read_parquet(TRAIN_PATH)
val_df   = pd.read_parquet(VAL_PATH)
print("Train:", len(train_df), " | Val:", len(val_df))

def build_text(df):
    return (df.get('subject', pd.Series([""]*len(df))).fillna('') + " " +
            df.get('body_text', pd.Series([""]*len(df))).fillna('') + " " +
            df.get('body_html', pd.Series([""]*len(df))).fillna('')).astype(str)

train_texts = build_text(train_df).tolist()
val_texts   = build_text(val_df).tolist()

train_labels = train_df['stage2_label'].astype(int).tolist()
val_labels   = val_df['stage2_label'].astype(int).tolist()

# ---------- TOKENIZER ----------
MODEL_NAME = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

def tok(batch):
    return tokenizer(batch["text"], truncation=True, padding="max_length", max_length=MAX_LEN)

train_ds = Dataset.from_dict({"text": train_texts, "labels": train_labels})
val_ds   = Dataset.from_dict({"text": val_texts,   "labels": val_labels})

train_ds = train_ds.map(tok, batched=True, remove_columns=["text"])
val_ds   = val_ds.map(tok,   batched=True, remove_columns=["text"])

train_ds.set_format("torch")
val_ds.set_format("torch")

data_collator = DataCollatorWithPadding(tokenizer)

# ---------- MODEL ----------
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2)

# CLASS WEIGHTS (numpy)
labels_np = np.array(train_labels + val_labels)
class_weights_np = compute_class_weight("balanced", classes=np.unique(labels_np), y=labels_np)
print("Class weights (numpy):", class_weights_np)

# ---------- CUSTOM TRAINER ----------
from transformers import Trainer as HFTrainer

class WeightedTrainer(HFTrainer):
    # accept extra args/kwargs to be compatible across transformers versions
    def compute_loss(self, model, inputs, return_outputs=False, *args, **kwargs):
        labels = inputs.get("labels")
        # move class weights to current device
        device = labels.device if labels is not None else (torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu"))
        class_w = torch.tensor(class_weights_np, dtype=torch.float, device=device)

        # forward
        outputs = model(**inputs)
        logits = outputs.get("logits")

        loss_fct = torch.nn.CrossEntropyLoss(weight=class_w)
        loss = loss_fct(logits.view(-1, model.config.num_labels), labels.view(-1))

        return (loss, outputs) if return_outputs else loss

# ---------- TRAINING ARGS ----------
training_args = TrainingArguments(
    output_dir=OUT_BERT,
    num_train_epochs=NUM_EPOCHS,
    per_device_train_batch_size=BATCH,
    per_device_eval_batch_size=BATCH*2,
    gradient_accumulation_steps=ACCUM_STEPS,
    learning_rate=LR,
    warmup_steps=10,
    weight_decay=0.01,
    fp16=torch.cuda.is_available(),
    logging_steps=20,
    save_steps=200,
    save_total_limit=2,
    seed=SEED,
    report_to="none"
)

# ---------- METRICS ----------
def compute_metrics(p):
    logits = p.predictions
    if isinstance(logits, tuple):
        logits = logits[0]
    preds = np.argmax(logits, axis=1)
    labels = p.label_ids
    pr, rc, f1, _ = precision_recall_fscore_support(labels, preds, average="binary", zero_division=0)
    acc = accuracy_score(labels, preds)
    return {"accuracy": acc, "precision": pr, "recall": rc, "f1": f1}

# ---------- TRAINER ----------
trainer = WeightedTrainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

print("🔥 Training DistilBERT Stage 2 (FP16 = {}) ...".format(training_args.fp16))
trainer.train()

# ---------- SAVE ----------
trainer.save_model(OUT_BERT)
tokenizer.save_pretrained(OUT_BERT)


print("TRAINING COMPLETE — model saved at:", OUT_BERT)

```

```python

# CELL 25 — Two-Stage Fusion (Stage1 + Stage2)
# - Load LightGBM
# - Load DistilBERT
# - Create final inference pipeline

import os, json, joblib, torch, numpy as np
from scipy.sparse import hstack, csr_matrix
from transformers import AutoTokenizer, AutoModelForSequenceClassification

BASE = "/content/drive/MyDrive/phish_project"
PROC = os.path.join(BASE, "processed")
MODEL_DIR = os.path.join(BASE, "models")

# ---------- Load Stage 1 artifacts ----------
lgb_path = os.path.join(MODEL_DIR, "lgb_fast_final.pkl")
tfidf_path = os.path.join(MODEL_DIR, "tfidf_stage1.pkl")
scaler_path = os.path.join(MODEL_DIR, "scaler_stage1.pkl")
meta_path = os.path.join(MODEL_DIR, "stage1_meta.json")

bst = joblib.load(lgb_path)
tfidf = joblib.load(tfidf_path)
scaler = joblib.load(scaler_path)
meta = json.load(open(meta_path))
num_cols_order = meta["numeric_feature_order"]

# ---------- Load Stage 2 (DistilBERT) ----------
BERT_DIR = os.path.join(MODEL_DIR, "distilbert_stage2")

tokenizer = AutoTokenizer.from_pretrained(BERT_DIR)
bert_model = AutoModelForSequenceClassification.from_pretrained(BERT_DIR)
bert_model.eval()
if torch.cuda.is_available():
    bert_model = bert_model.to("cuda")

# ---------- Numeric features extractors ----------
import re, json as _json

URL_RE = re.compile(r'(https?://[^\s"\']+)', re.I)

def compute_entropy(domain):
    from collections import Counter
    if not domain:
        return 0.0
    cnt = Counter(domain)
    probs = [v / len(domain) for v in cnt.values()]
    import math
    return -sum(p * math.log(p + 1e-9, 2) for p in probs)

def extract_features(urls_json, headers_json):
    # URLs
    try:
        if isinstance(urls_json, str):
            urls = _json.loads(urls_json)
            if not isinstance(urls, list):
                urls = []
        else:
            urls = []
    except:
        urls = []

    n_urls = len(urls)
    max_url_len = max((len(u) for u in urls), default=0)
    entropies = []
    any_short = 0
    any_ip = 0
    for u in urls:
        # entropy
        host = re.sub(r"^.*?//", "", u).split("/")[0]
        entropies.append(compute_entropy(host))
        # shortener?
        if any(x in u for x in ["bit.ly","t.co","tinyurl","goo.gl","is.gd","ow.ly"]):
            any_short = 1
        # raw IP?
        if re.search(r"\b\d{1,3}(?:\.\d{1,3}){3}\b", u):
            any_ip = 1
    mean_entropy = float(np.mean(entropies)) if entropies else 0.0

    # Headers
    try:
        h = _json.loads(headers_json) if isinstance(headers_json, str) else {}
    except:
        h = {}

    text = json.dumps(h).lower()
    spf_fail  = 1 if "spf=fail" in text else 0
    dkim_fail = 1 if "dkim=fail" in text else 0

    from_addr = h.get("From","") or h.get("from","")
    ret_path  = h.get("Return-Path","") or h.get("return-path","")

    def dom(addr):
        m = re.search(r"@([A-Za-z0-9.-]+)", str(addr))
        return m.group(1).lower() if m else ""

    mismatch = 1 if dom(from_addr) and dom(ret_path) and dom(from_addr) != dom(ret_path) else 0

    return {
        "n_urls": n_urls,
        "max_url_len": max_url_len,
        "mean_entropy": mean_entropy,
        "any_short": any_short,
        "any_ip": any_ip,
        "spf_fail": spf_fail,
        "dkim_fail": dkim_fail,
        "header_mismatch": mismatch,
    }

# ---------- Stage 1 prediction ----------
def stage1_predict(subject, body_text, body_html, urls_json, headers_json):
    full_text = f"{subject or ''} {body_text or ''} {body_html or ''}"
    X_text = tfidf.transform([full_text])

    feats = extract_features(urls_json, headers_json)
    X_num = scaler.transform([[feats[k] for k in num_cols_order]])

    X = hstack([X_text, csr_matrix(X_num)])
    p = float(bst.predict(X, num_iteration=bst.best_iteration or None)[0])
    return p, feats

# ---------- Stage 2 prediction (BERT) ----------
def stage2_predict(subject, body_text, body_html):
    text = f"{subject or ''} {body_text or ''} {body_html or ''}"
    enc = tokenizer(text, truncation=True, padding="max_length", max_length=256, return_tensors="pt")
    if torch.cuda.is_available():
        enc = {k: v.to("cuda") for k, v in enc.items()}
    with torch.no_grad():
        out = bert_model(**enc).logits
        probs = torch.softmax(out, dim=-1)[0].cpu().numpy()
    return float(probs[1])

# ---------- FINAL TWO-STAGE INFERENCE ----------
def two_stage_predict(email_dict):
    """
    email_dict = {
        'subject': str,
        'body_text': str,
        'body_html': str,
        'urls': '["http://..."]',
        'headers': '{"Return-Path": "..."}'
    }
    """
    s1_score, feats = stage1_predict(
        email_dict.get("subject"),
        email_dict.get("body_text"),
        email_dict.get("body_html"),
        email_dict.get("urls","[]"),
        email_dict.get("headers","{}")
    )

    # Decision: only run Stage2 if uncertain
    if 0.2 < s1_score < 0.8:
        s2_score = stage2_predict(
            email_dict.get("subject"),
            email_dict.get("body_text"),
            email_dict.get("body_html")
        )
        final_score = 0.4*s1_score + 0.6*s2_score
        label = int(final_score >= 0.6)
        return {
            "stage1_score": s1_score,
            "stage2_score": s2_score,
            "final_score": final_score,
            "label": label,
            "numeric_features": feats
        }

    # NO Stage2
    return {
        "stage1_score": s1_score,
        "stage2_score": None,
        "final_score": s1_score,
        "label": int(s1_score >= 0.6),
        "numeric_features": feats
    }


print(" TWO-STAGE FUSION MODULE LOADED SUCCESSFULLY ")
```

```python

# CELL 26 — Explainability (FIXED)
# - Safe shap usage for LightGBM (handles list output)
# - Avoids dangerous pip installs that conflict with numpy
# - Token attribution via Captum if available, else attention fallback


import os, json, joblib, numpy as np, pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

BASE = "/content/drive/MyDrive/phish_project"
PROC = os.path.join(BASE, "processed")
MODEL_DIR = os.path.join(BASE, "models")

# 0) Helpers: do not auto-install packages to avoid numpy conflicts
shap_available = True
try:
    import shap
except Exception as e:
    shap_available = False
    print("⚠ shap import failed or missing. Skipping SHAP plots. To enable SHAP, install a compatible version manually (note: numpy versions matter).")

# Captum optional
has_captum = True
try:
    from captum.attr import IntegratedGradients
except Exception:
    has_captum = False
    print("⚠ captum not available — will fallback to attention-based token scores.")

# 1) Load artifacts
lgb_path = os.path.join(MODEL_DIR, "lgb_fast_final.pkl")
tfidf_path = os.path.join(MODEL_DIR, "tfidf_stage1.pkl")
meta_path = os.path.join(MODEL_DIR, "stage1_meta.json")

bst = joblib.load(lgb_path)
tfidf = joblib.load(tfidf_path)
meta = json.load(open(meta_path))
num_cols = meta.get("numeric_feature_order", ["n_urls","max_url_len","mean_entropy","any_short","any_ip","spf_fail","dkim_fail","header_mismatch"])

# 2) Load df and ensure stage1_prob exists for sample
df_path = os.path.join(PROC, "emails_all_combined.parquet")
df = pd.read_parquet(df_path)
print("Loaded df rows:", len(df))

# If stage1_prob missing, compute it for the sample subset using tfidf + numeric scaler if scaler exists
scaler_path = os.path.join(MODEL_DIR, "scaler_stage1.pkl")
if 'stage1_prob' not in df.columns:
    print("stage1_prob missing — attempting to compute stage1_prob for all rows (this may be slow).")
    if os.path.exists(scaler_path):
        scaler = joblib.load(scaler_path)
        # ensure full_text
        if 'full_text' not in df.columns:
            df['full_text'] = df.get('subject','').fillna('') + ' ' + df.get('body_text','').fillna('') + ' ' + df.get('body_html','').fillna('')
        # build numeric features if not present
        missing_num = [c for c in num_cols if c not in df.columns]
        if missing_num:
            # try to use extract_features from session (Cell4), else set zeros
            try:
                from __main__ import extract_features
                num_rows = []
                for _, r in df.iterrows():
                    f = extract_features(r.get('urls','[]'), r.get('headers','{}'))
                    num_rows.append([f[c] for c in num_cols])
                X_num = np.array(num_rows)
            except Exception:
                print("Could not compute numeric features; filling zeros for numeric part.")
                X_num = np.zeros((len(df), len(num_cols)))
        else:
            X_num = df[num_cols].fillna(0).values
        # TF-IDF transform in chunks to avoid memory explosion
        batch_size = 20000
        probs = []
        for i in range(0, len(df), batch_size):
            chunk_text = df['full_text'].iloc[i:i+batch_size].astype(str).tolist()
            X_text_chunk = tfidf.transform(chunk_text)
            X_num_chunk = scaler.transform(X_num[i:i+batch_size])
            from scipy.sparse import hstack, csr_matrix
            X_chunk = hstack([X_text_chunk, csr_matrix(X_num_chunk)])
            p_chunk = bst.predict(X_chunk, num_iteration=bst.best_iteration or None)
            probs.extend(p_chunk.tolist())
        df['stage1_prob'] = probs
        print("Computed stage1_prob for all rows.")
    else:
        print("No scaler found at", scaler_path, "- cannot compute stage1_prob automatically. Please run Stage1 first.")

# 3) Build sample for SHAP / token attr
sample = df.sample(n=min(200, len(df)), random_state=42).reset_index(drop=True)

# 4) SHAP (numeric summary) — only if shap available
if shap_available:
    try:
        # build X_shap small dense matrix (safe size)
        X_text = tfidf.transform(sample['full_text'].fillna('').astype(str))
        if all([c in sample.columns for c in num_cols]):
            X_num = sample[num_cols].fillna(0).values
        else:
            # fallback compute numeric
            try:
                from __main__ import extract_features
                X_num = np.array([[extract_features(r.get('urls','[]'), r.get('headers','{}'))[k] for k in num_cols] if False else [extract_features(r.get('urls','[]'), r.get('headers','{}'))[k] for k in num_cols] for _,r in sample.iterrows()])
            except Exception:
                X_num = np.zeros((len(sample), len(num_cols)))
        from scipy.sparse import hstack, csr_matrix
        X_shap = hstack([X_text, csr_matrix(X_num)], format="csr")
        # convert to dense for shap (safe because sample small)
        X_shap_dense = X_shap[:200].toarray()
        print("Running SHAP TreeExplainer (this may take a few seconds)...")
        explainer = shap.TreeExplainer(bst)
        shap_values = explainer.shap_values(X_shap_dense)
        # shap_values can be list (one per class) or array
        if isinstance(shap_values, list) and len(shap_values) >= 2:
            sv = shap_values[1]   # class=1 (phish)
        else:
            sv = np.array(shap_values)
        # numeric part begins after text features
        n_text_feats = X_text.shape[1]
        if sv.ndim == 2 and sv.shape[1] > n_text_feats:
            numeric_shap = np.abs(sv[:, n_text_feats:]).mean(axis=0)
            feat_importance = sorted(zip(num_cols, numeric_shap), key=lambda x: x[1], reverse=True)
            print("Top numeric feature importances (SHAP):")
            for f, v in feat_importance:
                print(f, round(float(v), 6))
            # plot numeric importance
            plt.figure(figsize=(6,4))
            names = [f for f,_ in feat_importance]
            vals  = [v for _,v in feat_importance]
            plt.barh(range(len(names)), vals[::-1])
            plt.yticks(range(len(names)), names[::-1])
            plt.title("SHAP mean(|value|) — numeric features (Stage1)")
            plt.tight_layout()
            out_shap_numeric = os.path.join(MODEL_DIR, "shap_numeric_summary.png")
            plt.savefig(out_shap_numeric)
            plt.show()
            print("Saved numeric SHAP summary:", out_shap_numeric)
        else:
            print("SHAP output shape unexpected; skipping numeric summary.")
    except Exception as e:
        print("SHAP failed:", e)
else:
    print("Skipping SHAP because package is not available in this environment.")

# 5) Per-email numeric contributions (print top numeric contributions for a few samples)
print("\nSample per-email numeric SHAP (top contributions) — if SHAP was computed:")
if shap_available:
    try:
        # use sv if available
        if 'sv' in locals() and sv.ndim == 2:
            sample_susp = sample.sort_values('stage1_prob').head(2).append(sample.sort_values('stage1_prob').tail(3))
            for idx, row in sample_susp.iterrows():
                i = row.name
                # map sample idx to sv index (they align because we used sample)
                # find relative index
                rel_idx = i if i < sv.shape[0] else None
                if rel_idx is None:
                    continue
                numeric_part = sv[rel_idx, n_text_feats:]
                rec = sorted(list(zip(num_cols, numeric_part)), key=lambda x: abs(x[1]), reverse=True)[:10]
                print("\n=== Email idx:", i, "stage1_prob:", row.get('stage1_prob'))
                for k,v in rec:
                    print(k, round(float(v),6))
    except Exception as e:
        print("Per-email SHAP display failed:", e)

# 6) Token-level attribution for DistilBERT (Captum IG if present, else attention fallback)
from transformers import AutoTokenizer, AutoModelForSequenceClassification
bert_dir = os.path.join(MODEL_DIR, "distilbert_stage2")
tokenizer = AutoTokenizer.from_pretrained(bert_dir)
bert_model = AutoModelForSequenceClassification.from_pretrained(bert_dir)
bert_model.eval()
if torch.cuda.is_available():
    bert_model.to("cuda")

def token_attr(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding="max_length", max_length=256)
    if torch.cuda.is_available():
        inputs = {k:v.to("cuda") for k,v in inputs.items()}
    # Captum path
    if has_captum:
        try:
            ig = IntegratedGradients(bert_model)
            input_ids = inputs["input_ids"]
            att_mask = inputs.get("attention_mask", None)
            # define forward wrapper that returns phish prob
            def forward_func(input_ids, attention_mask=None):
                out = bert_model(input_ids=input_ids, attention_mask=attention_mask)
                probs = torch.softmax(out.logits, dim=-1)[:,1]
                return probs
            # baselines: zeros
            baseline = torch.zeros_like(input_ids)
            attr = ig.attribute(inputs=input_ids, baselines=baseline, additional_forward_args=(att_mask,), target=0)  # try target 0 or 1 depending; returns same shape as input_ids embeddings
            # sum attributions across embedding dim if needed (captum here may need embedding wrapper; this is a best-effort)
            try:
                scores = attr.squeeze(0).sum(dim=-1).detach().cpu().numpy()
            except Exception:
                scores = attr.squeeze(0).detach().cpu().numpy()
        except Exception as e:
            print("Captum IG failed, falling back to attention weights:", e)
            has_attn = True
            scores = None
    else:
        scores = None

    # Attention fallback if IG not available or failed
    if scores is None:
        with torch.no_grad():
            out = bert_model(**inputs, output_attentions=True)
            attns = out.attentions  # tuple of (layers)
            # average heads & layers, sum importance per token
            att_stack = torch.stack(attns)  # layers x batch x heads x seq x seq
            att_mean = att_stack.mean(dim=0).mean(dim=1).squeeze(0)  # seq x seq
            token_scores = att_mean.sum(dim=0).cpu().numpy()
            scores = token_scores

    toks = tokenizer.convert_ids_to_tokens(inputs["input_ids"].squeeze(0).cpu().numpy())
    return list(zip(toks, scores))

# Example: compute token attribution for one sample and save CSV
example_text = (sample.iloc[0]['subject'] or '') + " " + (sample.iloc[0]['body_text'] or '')
tok_attr = token_attr(example_text)
print("\nToken attribution (example):", tok_attr[:30])
pd.DataFrame(tok_attr, columns=["token","score"]).to_csv(os.path.join(MODEL_DIR,"bert_token_attr_sample.csv"), index=False)
print("Saved token attribution CSV.")
```

```python
# Cell 28
#!/usr/bin/env python3
"""
evaluate_comparison.py

Script đánh giá Ablation Study (S1 vs S2 vs Fusion) và Data Drift.
"""
import argparse
import os
import json
import joblib
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score,
    roc_curve, precision_recall_curve, average_precision_score, precision_score,
    recall_score, f1_score, accuracy_score
)
import matplotlib.pyplot as plt
import seaborn as sns
import torch
from scipy.sparse import hstack, csr_matrix
import sys
import lightgbm as lgb
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split # Cần cho Data Drift Retraining

# HF transformers
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# -----------------------
# Helpers (Tái định nghĩa để script độc lập)
# -----------------------
def ensure_dir(p):
    Path(p).mkdir(parents=True, exist_ok=True)

import re, json as _json, math
from collections import Counter # Cần cho compute_entropy

def compute_entropy(domain):
    if not domain:
        return 0.0
    cnt = Counter(domain)
    probs = [v/len(domain) for v in cnt.values()]
    return -sum(p*math.log(p+1e-9, 2) for p in probs)

def extract_features_from_row(urls_json, headers_json, num_cols_order):
    """Trích xuất 8 numeric features từ JSON đầu vào."""
    try:
        urls = _json.loads(urls_json) if isinstance(urls_json, str) else []
        if not isinstance(urls, list): urls = []
    except Exception: urls = []
    n_urls = len(urls)
    max_url_len = max((len(u) for u in urls), default=0)
    entropies = []; any_short = 0; any_ip = 0
    for u in urls:
        host = re.sub(r"^.*?//", "", u).split("/")[0]
        entropies.append(compute_entropy(host))
        if any(x in u for x in ["bit.ly","t.co","tinyurl","goo.gl","is.gd","ow.ly"]): any_short = 1
        if re.search(r"\b\d{1,3}(?:\.\d{1,3}){3}\b", u): any_ip = 1
    mean_entropy = float(np.mean(entropies)) if entropies else 0.0

    try: h = _json.loads(headers_json) if isinstance(headers_json, str) else {}
    except: h = {}
    txt = json.dumps(h).lower()
    spf_fail = 1 if "spf=fail" in txt else 0
    dkim_fail = 1 if "dkim=fail" in txt else 0

    from_addr = h.get("From","") or h.get("from","")
    ret_path  = h.get("Return-Path","") or h.get("return-path","")
    def dom(addr):
        m = re.search(r"@([A-Za-z0-9.-]+)", str(addr))
        return m.group(1).lower() if m else ""
    mismatch = 1 if dom(from_addr) and dom(ret_path) and dom(from_addr) != dom(ret_path) else 0

    feats = {
        "n_urls": n_urls, "max_url_len": max_url_len, "mean_entropy": mean_entropy,
        "any_short": any_short, "any_ip": any_ip, "spf_fail": spf_fail,
        "dkim_fail": dkim_fail, "header_mismatch": mismatch
    }
    return [feats[c] for c in num_cols_order]

# -----------------------
# Load Models
# -----------------------
def load_models(model_dir):
    """Tải tất cả artifacts đã lưu, trả về dict."""
    res = {}
    try:
        res['lgb'] = joblib.load(Path(model_dir) / "lgb_fast_final.pkl")
        res['tfidf'] = joblib.load(Path(model_dir) / "tfidf_stage1.pkl")
        res['scaler'] = joblib.load(Path(model_dir) / "scaler_stage1.pkl")
        res['meta'] = json.load(open(Path(model_dir) / "stage1_meta.json"))
        print("Loaded LGB and Stage1 preprocessors.")
    except Exception as e:
        print(f"Error loading Stage1 artifacts: {e}")
        res['lgb'], res['tfidf'], res['scaler'] = None, None, None

    try:
        bert_dir = Path(model_dir) / "distilbert_stage2"
        if bert_dir.exists():
            res['bert_tokenizer'] = AutoTokenizer.from_pretrained(str(bert_dir))
            res['bert_model'] = AutoModelForSequenceClassification.from_pretrained(str(bert_dir))
            res['bert_model'].eval()
            if torch.cuda.is_available():
                res['bert_model'] = res['bert_model'].to("cuda")
            print("Loaded DistilBERT stage2.")
        else:
            print("DistilBERT dir not found.")
    except Exception as e:
        print(f" Error loading DistilBERT: {e}")

    return res

# -----------------------
# Predictions wrappers
# -----------------------
def predict_stage1_batch(df, models):
    """Trả về xác suất S1 (LGB) trên DataFrame."""
    if models['lgb'] is None: raise RuntimeError("Stage1 model not loaded.")
    tfidf = models['tfidf']
    scaler = models['scaler']
    num_cols = models['meta']['numeric_feature_order'] if models['meta'] else ["n_urls","max_url_len","mean_entropy","any_short","any_ip","spf_fail","dkim_fail","header_mismatch"]

    df['full_text_concat'] = (df.get('subject','').fillna('') + " " + df.get('body_text','').fillna('') + " " + df.get('body_html','').fillna('')).astype(str)
    texts = df['full_text_concat'].tolist()

    X_text = tfidf.transform(texts)

    num_rows = []
    for _,r in df.iterrows():
        row_vals = extract_features_from_row(r.get('urls','[]'), r.get('headers','{}'), num_cols)
        num_rows.append(row_vals)
    X_num = scaler.transform(np.array(num_rows))

    X = hstack([X_text, csr_matrix(X_num)])
    probs = models['lgb'].predict(X, num_iteration=models['lgb'].best_iteration or None)
    return np.array(probs)

def predict_stage2_batch(df, models, batch_size=64):
    """Trả về xác suất S2 (BERT) trên DataFrame."""
    tokenizer = models.get('bert_tokenizer')
    model = models.get('bert_model')
    if tokenizer is None or model is None: raise RuntimeError("Stage2 model not loaded.")

    df['full_text_concat'] = (df.get('subject','').fillna('') + " " + df.get('body_text','').fillna('') + " " + df.get('body_html','').fillna('')).astype(str)
    texts = df['full_text_concat'].tolist()

    probs = []
    device = "cuda" if torch.cuda.is_available() else "cpu"

    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i+batch_size]
        enc = tokenizer(batch_texts, truncation=True, padding=True, max_length=256, return_tensors="pt")
        enc = {k:v.to(device) for k,v in enc.items()}
        with torch.no_grad():
            out = model(**enc).logits
            p = torch.softmax(out, dim=-1)[:,1].detach().cpu().numpy()
            probs.extend(p.tolist())
    return np.array(probs)

def two_stage_predict_batch(df, models, low_threshold=0.2, high_threshold=0.8):
    # Stage1 probs
    s1 = predict_stage1_batch(df, models)
    final = s1.copy()
    # prepare s2 array full of NaN
    s2 = np.full_like(s1, fill_value=np.nan, dtype=float)

    # mask những mẫu cần gọi Stage2
    mask = (s1 > low_threshold) & (s1 < high_threshold)
    if mask.any():
        df_mask = df[mask].reset_index(drop=True)
        try:
            s2_vals = predict_stage2_batch(df_mask, models)
        except Exception as e:
            print(" Error when predicting Stage2 for masked samples:", e)
            s2_vals = np.array([np.nan]*mask.sum())
        # assign s2 values into proper positions
        s2[mask] = s2_vals
        # fuse final score where s2 present (otherwise keep s1)
        final[mask] = 0.4 * s1[mask] + 0.6 * s2_vals
        print(f"✨ Fusion applied to {mask.sum()} samples.")
    else:
        print("ℹ No samples passed to Stage2 (all predicted confident by Stage1).")

    return s1, s2, final

# -----------------------
# Main evaluation flow
# -----------------------
def evaluate_all(args):
    ensure_dir(args.out_dir)
    models = load_models(args.model_dir)

    # Load data
    df = pd.read_parquet(args.data)
    if 'label' not in df.columns: raise RuntimeError("No 'label' column in data.")
    y_true_all = df['label'].astype(int).values

    # 1. Định nghĩa Evaluation Set (Lưu ý: Không chia train/test ngẫu nhiên trong script này,
    # mà dùng toàn bộ data hoặc filter. Nếu muốn dùng Test set đã chia, cần load riêng.)
    if args.test_filter is not None:
        mask = df.eval(args.test_filter)
        df_eval = df[mask].reset_index(drop=True)
    else:
        # Nếu không có filter, dùng toàn bộ data
        df_eval = df.reset_index(drop=True)

    print(f"Evaluation set size: {len(df_eval)} (Phish: {df_eval['label'].sum()})")
    y_true = df_eval['label'].astype(int).values

    # 1) Stage1-only
    if models['lgb'] is not None:
        s1_probs = predict_stage1_batch(df_eval, models)
        s1_preds = (s1_probs >= args.stage1_thresh).astype(int)
        compute_metrics_and_save(y_true, s1_preds, s1_probs, "stage1_only", args.out_dir)

    # 2) Stage2-only (Ablation: BERT chạy trên tất cả)
    if models.get('bert_model') is not None:
        try:
            s2_probs = predict_stage2_batch(df_eval, models, batch_size=args.bert_batch)
            s2_preds = (s2_probs >= args.stage2_thresh).astype(int)
            compute_metrics_and_save(y_true, s2_preds, s2_probs, "stage2_only", args.out_dir)
        except Exception as e:
            print(f"Stage2-only evaluation failed: {e}")

    # 3) Two-Stage fusion
    if models['lgb'] is not None and models.get('bert_model') is not None:
        s1, s2_vals, final = two_stage_predict_batch(df_eval, models, low_threshold=args.low_threshold, high_threshold=args.high_threshold)
        final_probs = final
        final_preds = (final_probs >= args.final_thresh).astype(int)
        compute_metrics_and_save(y_true, final_preds, final_probs, "two_stage", args.out_dir)

        # FN Reduction Analysis
        s1_preds = (s1 >= args.stage1_thresh).astype(int)
        fn_s1 = int(((y_true==1) & (s1_preds==0)).sum())
        fn_two = int(((y_true==1) & (final_preds==0)).sum())
        reduction = fn_s1 - fn_two
        reduction_pct = (reduction / fn_s1 * 100.0) if fn_s1>0 else None

        fn_res = {"fn_stage1": fn_s1, "fn_two_stage": fn_two, "fn_reduction_pct": reduction_pct}
        json.dump(fn_res, open(Path(args.out_dir)/"fn_reduction.json", "w"), indent=2)
        print(f"Saved FN reduction: {fn_res}")

    # 4) Drift experiment (Tùy chọn)
    # ... (Logic Data Drift như đã định nghĩa) ...

    print("Evaluation complete. Results & plots are in:", args.out_dir)

# -----------------------
# CLI (Chạy bằng sys.argv)
# -----------------------
def parse_args():
    p = argparse.ArgumentParser()
    # (Định nghĩa các tham số như code gốc)
    p.add_argument("--data", required=True, help="Path to emails_all_combined.parquet")
    p.add_argument("--model_dir", required=True, help="models directory")
    p.add_argument("--out_dir", required=True, help="output directory to store metrics & plots")
    p.add_argument("--test_filter", default=None)
    p.add_argument("--stage1_thresh", type=float, default=0.6)
    p.add_argument("--stage2_thresh", type=float, default=0.5)
    p.add_argument("--final_thresh", type=float, default=0.6)
    p.add_argument("--low_threshold", type=float, default=0.2)
    p.add_argument("--high_threshold", type=float, default=0.8)
    p.add_argument("--bert_batch", type=int, default=32)
    p.add_argument("--date_col", default="date")
    p.add_argument("--drift_old_cutoff", default=None)
    p.add_argument("--drift_new_cutoff", default=None)
    p.add_argument("--retrain_stage1", action="store_true")
    p.add_argument("--tfidf_max_features", type=int, default=150000)

    return p.parse_args()

def compute_metrics_and_save(y_true, y_pred, y_prob, out_prefix, out_dir):
    ensure_dir(out_dir)
    # logic metrics and plotting
    report = classification_report(y_true, y_pred, output_dict=True, digits=4)
    cm = confusion_matrix(y_true, y_pred)
    acc = accuracy_score(y_true, y_pred)
    pr = precision_score(y_true, y_pred, zero_division=0)
    rc = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    metrics = {"report": report, "confusion_matrix": cm.tolist(), "accuracy": acc, "precision": pr, "recall": rc, "f1": f1}
    json.dump(metrics, open(Path(out_dir)/f"{out_prefix}_metrics.json", "w"), indent=2)

    # Plotting (ROC, PR, ConfMat)
    # ... (Logic plotting đã được định nghĩa trong code gốc) ...

    # Example plot ConfMat
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=[0,1], yticklabels=[0,1])
    plt.title(f"ConfMat: {out_prefix}")
    plt.savefig(Path(out_dir)/f"{out_prefix}_confmat.png")
    plt.close()

    print(f"✅ Saved metrics JSON and plots for {out_prefix}")
    return metrics


# -----------------------
# MAIN EXECUTION BLOCK
# -----------------------
if __name__ == "__main__":
    # ----------------- ĐỊNH NGHĨA THAM SỐ TRỰC TIẾP (GIẢ LẬP CLI) -----------------
    DATA_PATH = "/content/drive/MyDrive/phish_project/processed/emails_all_combined.parquet"
    MODEL_PATH = "/content/drive/MyDrive/phish_project/models"
    OUTPUT_PATH = "/content/drive/MyDrive/phish_project/models/analysis"

    # Giả lập tham số CLI
    sys.argv = [
        'evaluate_comparison.py',
        '--data', DATA_PATH,
        '--model_dir', MODEL_PATH,
        '--out_dir', OUTPUT_PATH,
        '--final_thresh', '0.6',
        '--low_threshold', '0.2',
        '--high_threshold', '0.8',
        '--bert_batch', '64',
        '--date_col', 'date',
        '--drift_old_cutoff', '2021-01-01',
        '--drift_new_cutoff', '2024-01-01',
        # Thêm cờ --retrain_stage1 nếu muốn chạy lại LGB trên tập cũ cho drift
    ]
    # -------------------------------------------------------------------------------

    print("🚀 Bắt đầu chạy Ablation Study và Data Drift Evaluation (sử dụng sys.argv)...")
    try:
        args = parse_args()
        evaluate_all(args)
    except SystemExit:
        print("Đánh giá thất bại: Kiểm tra lại các tham số trong sys.argv.")
    except Exception as e:
        print(f"Đánh giá thất bại với lỗi: {e}")
```

# Mở rộng & cân bằng Stage2 (hybrid top-K uncertain + FN candidates)

```python
# === Cell 1: stage2_hybrid_sampling (ĐÃ SỬA LỖI PATH) === không cần chạy lại
import numpy as np
import pandas as pd
import os
import random
import math
from sklearn.utils import resample

# --- SỬA LẠI ĐƯỜNG DẪN NÀY CHO KHỚP VỚI CELL TRƯỚC ---
PROC = "/content/drive/MyDrive/phish_project/processed"
# -----------------------------------------------------

OUT_SUS = os.path.join(PROC, "emails_suspicious_set.parquet")

# Kiểm tra file có tồn tại không trước khi đọc
if not os.path.exists(OUT_SUS):
    print(f"❌ Lỗi: Không tìm thấy file tại {OUT_SUS}")
    print("👉 Hãy chắc chắn bạn đã chạy Cell 18 (Lọc Suspicious Set) thành công.")
else:
    df = pd.read_parquet(OUT_SUS)
    print("Loaded suspicious:", len(df))

    # Target tổng mẫu stage2 (tùy chỉnh)
    TARGET_N = 3000   # đặt 2000-5000 nếu muốn
    MIN_PER_CLASS = 300  # tối thiểu/label nếu stratify

    # 1) Top-K uncertain (near 0.5) => lấy tỉ lệ lớn
    if 'stage1_prob' in df.columns:
        df['uncertainty'] = np.abs(df['stage1_prob'] - 0.5)
        topk = df.sort_values('uncertainty').head(int(TARGET_N * 0.7)).copy()  # 70% từ uncertain
    else:
        print("⚠ Không thấy cột stage1_prob, lấy mẫu ngẫu nhiên.")
        topk = df.sample(min(len(df), int(TARGET_N * 0.7)))

    # 2) FN-candidates heuristics
    suspicious_keywords = ["verify", "account", "login", "password", "click", "update", "confirm", "invoice", "urgent"]
    def has_suspicious_kw(text):
        t = str(text).lower()
        return any(k in t for k in suspicious_keywords)

    if 'stage1_prob' in df.columns:
        fn_cands = df[(df['stage1_prob'] < 0.3) & df['full_text'].apply(has_suspicious_kw)].copy()
    else:
        fn_cands = df[df['full_text'].apply(has_suspicious_kw)].copy()

    fn_take = fn_cands.sample(min(len(fn_cands), int(TARGET_N * 0.3)), random_state=42)  # 30% from FN candidates

    # 3) Merge and dedup
    stage2_pool = pd.concat([topk, fn_take])
    # Chỉ drop duplicates nếu các cột này tồn tại
    cols_to_check = [c for c in ["message_id", "full_text", "raw_email"] if c in stage2_pool.columns]
    if cols_to_check:
        stage2_pool = stage2_pool.drop_duplicates(subset=cols_to_check)

    stage2_pool = stage2_pool.sample(frac=1, random_state=42).reset_index(drop=True)
    print("Stage2 pool size:", len(stage2_pool))

    # If < TARGET_N, optionally relax thresholds / add random uncertain from outside window
    if len(stage2_pool) < TARGET_N and 'uncertainty' in df.columns:
        need = TARGET_N - len(stage2_pool)
        # Lấy thêm các mẫu chưa có trong stage2_pool
        remaining = df[~df.index.isin(stage2_pool.index)]
        if len(remaining) > 0:
            extra = remaining.sort_values('uncertainty').head(need)
            stage2_pool = pd.concat([stage2_pool, extra]).reset_index(drop=True)
            print("Padded with extra uncertain ->", len(stage2_pool))

    # Ensure class balance suggestion: mark as 'to_review' for manual labeling if no auto labels
    if 'stage2_label' not in stage2_pool.columns:
        stage2_pool['stage2_label'] = np.nan

    stage2_pool_path = os.path.join(PROC, "stage2_pool_for_annotation.parquet")
    stage2_pool.to_parquet(stage2_pool_path, index=False)
    print("Saved stage2 pool ->", stage2_pool_path)
```

```python
# [Cell A - FIXED] Mining dữ liệu khó từ tập GỐC để đạt target 3000 mẫu
import numpy as np
import pandas as pd
import os

# Cấu hình đường dẫn
PROC = "/content/drive/MyDrive/phish_project/processed"
IN_FULL = os.path.join(PROC, "emails_all_combined.parquet") # Đọc file gốc
OUT_POOL = os.path.join(PROC, "stage2_train_expanded.parquet")

TARGET_N = 3000  # Mục tiêu

print(f" Đang đọc tập dữ liệu gốc từ: {IN_FULL}")
if not os.path.exists(IN_FULL):
    print(" Lỗi: Không tìm thấy file gốc. Hãy kiểm tra lại đường dẫn.")
else:
    df_all = pd.read_parquet(IN_FULL)
    print(f"   - Tổng số email gốc: {len(df_all)}")

    # 1. Nếu chưa có cột 'stage1_prob', cần load lại model Stage 1 để dự đoán (QUAN TRỌNG)
    # Tuy nhiên, để tiết kiệm thời gian, ta sẽ dùng chiến lược Heuristic (Từ khóa + Độ dài)
    # nếu cột 'stage1_prob' bị thiếu.

    if 'stage1_prob' in df_all.columns:
        print("   - Tìm thấy cột 'stage1_prob', sử dụng Uncertainty Sampling...")
        df_all['uncertainty'] = np.abs(df_all['stage1_prob'] - 0.5)
        # Lấy 2000 mẫu "khó nhất" (gần 0.5 nhất)
        df_hard = df_all.sort_values('uncertainty').head(2000)
    else:
        print("⚠ Không thấy 'stage1_prob', dùng chiến lược lấy mẫu ngẫu nhiên có trọng số (Heuristics)...")
        # Lấy các email có chứa keyword nhạy cảm nhưng có thể bị sót
        keywords = ["verify", "account", "urgent", "update", "bank", "invoice", "security"]
        def has_kw(t): return any(k in str(t).lower() for k in keywords)

        df_risk = df_all[df_all['full_text'].apply(has_kw)]
        if len(df_risk) > 2000:
            df_hard = df_risk.sample(2000, random_state=42)
        else:
            df_hard = df_risk

    # 2. Lấy thêm các mẫu Phishing (Label=1) để đảm bảo model học được dấu hiệu tấn công
    # Lấy 1000 mẫu Phishing ngẫu nhiên từ tập gốc
    df_phish = df_all[df_all['label'] == 1]
    if len(df_phish) > 0:
        take_n = min(len(df_phish), 1000)
        df_phish_sample = df_phish.sample(take_n, random_state=42)
    else:
        df_phish_sample = pd.DataFrame()

    # 3. Gộp lại
    df_pool = pd.concat([df_hard, df_phish_sample]).drop_duplicates(subset=['raw_hash'])

    # Nếu vẫn chưa đủ 3000, lấy thêm ngẫu nhiên từ tập gốc (Label 0 và 1)
    if len(df_pool) < TARGET_N:
        needed = TARGET_N - len(df_pool)
        print(f"   - Còn thiếu {needed} mẫu, lấy thêm ngẫu nhiên...")
        remaining = df_all[~df_all.index.isin(df_pool.index)]
        if len(remaining) > 0:
            df_extra = remaining.sample(min(len(remaining), needed), random_state=42)
            df_pool = pd.concat([df_pool, df_extra])

    # Shuffle và lưu
    df_pool = df_pool.sample(frac=1, random_state=42).reset_index(drop=True)

    # Đảm bảo cột label tồn tại và đúng kiểu
    df_pool['label'] = df_pool['label'].fillna(0).astype(int)
    # Stage 2 cần cột 'stage2_label', ta gán tạm bằng label gốc
    df_pool['stage2_label'] = df_pool['label']

    print("-" * 30)
    print(f" Đã tạo tập train Stage 2 MỚI: {len(df_pool)} mẫu")
    print(f"   - Phân bố nhãn: {df_pool['label'].value_counts().to_dict()}")

    df_pool.to_parquet(OUT_POOL, index=False)
    print(f" Đã lưu tại: {OUT_POOL}")
    print(" file này cho bước Data Augmentation .")
```

```python
# === Cell B: augmentation_paraphrase_backtranslation ===
!pip install -q transformers sentencepiece accelerate --upgrade
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch, random

# Config
AUG_PER_SAMPLE = 2   # mỗi email tạo 0..2 biến thể
TARGET_AUG = 2000     # tổng augment tối đa

# Load small T5 paraphrase (tùy chọn: "Vamsi/T5_Paraphrase_Paws")
try:
    tok = AutoTokenizer.from_pretrained("Vamsi/T5_Paraphrase_Paws")
    model = AutoModelForSeq2SeqLM.from_pretrained("Vamsi/T5_Paraphrase_Paws")
    model = model.to("cuda" if torch.cuda.is_available() else "cpu")
    use_paraphrase = True
except Exception as e:
    print("Paraphrase model load failed:", e)
    use_paraphrase = False

def paraphrase(text, n=1, max_len=256):
    if not use_paraphrase: return []
    prefix = "paraphrase: "
    inputs = tok(prefix + text, return_tensors="pt", truncation=True, max_length=max_len).to(model.device)
    outs = model.generate(**inputs, num_beams=5, num_return_sequences=n, max_length=max_len)
    return [tok.decode(o, skip_special_tokens=True, clean_up_tokenization_spaces=True) for o in outs]

# Back-translation (en->fr->en) example using Helsinki small models (might be slower)
try:
    t_en_fr = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-fr").to(model.device)
    tok_en_fr = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-fr")
    t_fr_en = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-fr-en").to(model.device)
    tok_fr_en = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-fr-en")
    use_bt = True
except Exception as e:
    print("Back-translation load failed:", e)
    use_bt = False

def back_translate(text):
    if not use_bt: return None
    i = tok_en_fr(text, return_tensors="pt", truncation=True, max_length=256).to(model.device)
    mid = t_en_fr.generate(**i, max_length=256, num_beams=4)
    midtxt = tok_en_fr.decode(mid[0], skip_special_tokens=True)
    j = tok_fr_en(midtxt, return_tensors="pt", truncation=True, max_length=256).to(model.device)
    out = t_fr_en.generate(**j, max_length=256, num_beams=4)
    return tok_fr_en.decode(out[0], skip_special_tokens=True)

# Apply augmentation to a subset (only to suspicious pool to avoid label drift)
stage2_pool = pd.read_parquet(os.path.join(PROC, "stage2_pool_for_annotation.parquet"))
augmented = []
for i,row in stage2_pool.sample(frac=0.5, random_state=42).iterrows():  # sample fraction to limit cost
    text = str(row['full_text'])
    # paraphrase
    for p in paraphrase(text, n=min(2,AUG_PER_SAMPLE)):
        augmented.append({**row.to_dict(), 'full_text': p, 'aug_type':'paraphrase'})
    # back-translation one-shot
    bt = back_translate(text)
    if bt:
        augmented.append({**row.to_dict(), 'full_text': bt, 'aug_type':'back_translation'})
    if len(augmented) >= TARGET_AUG: break

aug_df = pd.DataFrame(augmented)
print("Created augmentations:", len(aug_df))
if len(aug_df)>0:
    aug_path = os.path.join(PROC, "stage2_augmentations.parquet")
    aug_df.to_parquet(aug_path, index=False)
    print("Saved aug ->", aug_path)
```

```python
# [Cell C - MAX POWER] Tạo Adversarial Dataset số lượng lớn (Mass Generation)
import re
import random
import pandas as pd
import os
from tqdm.notebook import tqdm

# Cấu hình đường dẫn
PROC = "/content/drive/MyDrive/phish_project/processed"

# Dùng file Expanded (3000 mẫu) làm nguồn để đảm bảo đủ dữ liệu gốc
IN_FILE = os.path.join(PROC, "stage2_train_expanded.parquet")
OUT_FILE = os.path.join(PROC, "stage2_adversarial_samples.parquet")

# --- Các hàm tạo mẫu tấn công (Giữ nguyên logic) ---
def misspell_brand(text, brand_list=None):
    if brand_list is None:
        brand_list = ["paypal","amazon","microsoft","google","apple","facebook","netflix", "verify", "security", "update", "bank"]
    t = str(text)
    # Tăng xác suất gây nhiễu lên một chút để tạo dữ liệu khó hơn
    changed = False
    for b in brand_list:
        pattern = re.compile(re.escape(b), flags=re.I)
        if pattern.search(t):
            def repl(m):
                s = m.group(0)
                # Thay thế ký tự homoglyphs
                if random.random() < 0.6: # 60% cơ hội bị thay thế
                    s2 = s.replace('o','0').replace('l','1').replace('i','!').replace('a','@').replace('e','3').replace('s','$')
                else:
                    # Xóa hoặc nhân đôi ký tự
                    if len(s) > 2:
                        idx = random.randint(1, len(s)-1)
                        if random.random() < 0.5:
                            s2 = s[:idx] + s[idx+1:] # Xóa
                        else:
                            s2 = s[:idx] + s[idx] + s[idx:] # Nhân đôi
                    else:
                        s2 = s
                return s2
            t_new = pattern.sub(repl, t)
            if t_new != t:
                changed = True
                t = t_new

    # Nếu không tìm thấy brand nào, random gây nhiễu một từ dài bất kỳ
    if not changed:
        words = t.split()
        if len(words) > 5:
            idx = random.randint(0, len(words)-1)
            if len(words[idx]) > 4:
                words[idx] = words[idx].replace('a','@').replace('o','0')
            t = " ".join(words)
    return t

def shorten_urls(text):
    # Thay thế MỌI link tìm thấy
    return re.sub(r'https?://[^\s\'">]+', lambda m: "https://bit.ly/{}".format(''.join(random.choices('abcdefghijklmnopqrstuvwxyz0123456789', k=7))), str(text))

def inject_html(text):
    # Chèn nhiều rác hơn
    inj = '<div style="display:none; opacity:0">System Check: OK</div>'
    hid = ''
    return inj + str(text) + hid

# --- Main Logic Tăng Số Lượng ---
if os.path.exists(IN_FILE):
    print(f" Đang đọc nguồn dữ liệu từ: {IN_FILE}")
    stage2 = pd.read_parquet(IN_FILE)

    # Lấy tối đa 2500 mẫu gốc để nhân bản (tránh quá tải RAM nếu file quá lớn)
    if len(stage2) > 2500:
        stage2 = stage2.sample(2500, random_state=42)

    adv_rows = []
    print(f"Bắt đầu nhân bản dữ liệu (Mass Generation) từ {len(stage2)} mẫu gốc...")

    for _, r in tqdm(stage2.iterrows(), total=len(stage2)):
        base = r.to_dict()
        original_text = str(base.get('full_text', ''))

        # Chiến thuật: Mỗi email gốc tạo ra 3 biến thể tấn công khác nhau

        # 1. Biến thể 1: Lỗi chính tả (Typosquatting) -> Luôn tạo
        adv_rows.append({
            **base,
            'full_text': misspell_brand(original_text),
            'adv_type': 'misspell_brand'
        })

        # 2. Biến thể 2: HTML Injection -> Luôn tạo
        adv_rows.append({
            **base,
            'full_text': inject_html(original_text),
            'adv_type': 'injected_html'
        })

        # 3. Biến thể 3: URL Shortener -> Chỉ tạo nếu có link
        if 'http' in original_text:
            adv_rows.append({
                **base,
                'full_text': shorten_urls(original_text),
                'adv_type': 'url_shortener'
            })

    adv_df = pd.DataFrame(adv_rows)
    # Loại bỏ dòng trùng hoặc rỗng
    adv_df = adv_df.drop_duplicates(subset=['full_text'])
    adv_df = adv_df[adv_df['full_text'].str.len() > 5]

    adv_df.to_parquet(OUT_FILE, index=False)
    print("-" * 30)
    print(f" Đã tạo xong tập Adversarial !")
    print(f"   - Tổng số mẫu: {len(adv_df)} (Tăng gấp nhiều lần so với 219)")
    print(f"   - Lưu tại: {OUT_FILE}")
else:
    print(f" Lỗi: Không tìm thấy file {IN_FILE}. Bạn cần chạy Cell A (Fixed) trước.")
```

```python
# [Cell D - FIXED] Đánh giá Robustness trên tập Adversarial (Đã sửa đường dẫn Model)
import time
import pandas as pd
import os
import torch
import numpy as np
from sklearn.metrics import classification_report, accuracy_score
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# --- CẤU HÌNH ĐƯỜNG DẪN CHÍNH XÁC ---
BASE = "/content/drive/MyDrive/phish_project"
PROC = os.path.join(BASE, "processed")
MODEL_DIR = os.path.join(BASE, "models")

# File dữ liệu tấn công
ADV_FILE = os.path.join(PROC, "stage2_adversarial_samples.parquet")
# Thư mục chứa model DistilBERT đã train
BERT_DIR = os.path.join(MODEL_DIR, "distilbert_stage2")

# --- KIỂM TRA TỒN TẠI ---
if not os.path.exists(ADV_FILE):
    print(f" Lỗi: Không tìm thấy file dữ liệu {ADV_FILE}. Hãy chạy Cell C trước.")
elif not os.path.exists(BERT_DIR):
    print(f" Lỗi: Không tìm thấy thư mục model {BERT_DIR}.")
    print("Bạn đã chạy Cell 24 (Training DistilBERT) chưa? Hãy quay lại train model trước.")
else:
    print(f" Đang tải dữ liệu từ: {ADV_FILE}")
    adv_df = pd.read_parquet(ADV_FILE)

    print(f" Đang tải model từ: {BERT_DIR}")
    try:
        tokenizer = AutoTokenizer.from_pretrained(BERT_DIR)
        model = AutoModelForSequenceClassification.from_pretrained(BERT_DIR)
        device = "cuda" if torch.cuda.is_available() else "cpu"
        model = model.to(device)
        model.eval()

        def predict_texts(texts, batch=32):
            preds = []
            probs = []

            # Chia nhỏ batch để tránh tràn RAM
            for i in range(0, len(texts), batch):
                batch_texts = list(texts[i:i+batch])
                inputs = tokenizer(batch_texts, return_tensors="pt", truncation=True, padding=True, max_length=256).to(device)

                with torch.no_grad():
                    out = model(**inputs)
                    logits = out.logits.cpu().numpy()
                    # Softmax để ra xác suất
                    p = 1 / (1 + np.exp(-logits))  # Sigmoid nếu binary output 1 logit
                    if logits.shape[1] > 1:
                        # Nếu output có 2 logits (0 và 1), lấy cột 1 và softmax
                        prob_tensor = torch.softmax(torch.tensor(logits), dim=1)
                        p = prob_tensor[:, 1].numpy()
                    else:
                        p = p[:, 0]

                    preds += list((p > 0.5).astype(int))
                    probs += list(p)
            return preds, probs

        print(f"🚀 Bắt đầu đánh giá trên {len(adv_df)} mẫu tấn công...")
        adv_texts = adv_df['full_text'].astype(str).tolist()

        # Chạy dự đoán
        preds, probs = predict_texts(adv_texts, batch=32)
        adv_df['pred'] = preds
        adv_df['pred_prob'] = probs

        # Nếu có label gốc (thường Cell C đã copy label sang), ta tính metrics
        # Lưu ý: Cell C dùng 'label' hoặc 'stage2_label', ta check cả hai
        true_label_col = 'label' if 'label' in adv_df.columns else ('stage2_label' if 'stage2_label' in adv_df.columns else None)

        if true_label_col:
            y_true = adv_df[true_label_col].fillna(1).astype(int) # Fillna 1 vì adversarial thường là phish
            y_pred = adv_df['pred'].astype(int)


            print(" KẾT QUẢ ROBUSTNESS TEST")

            print(classification_report(y_true, y_pred, digits=4))
            print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")

            # Save failure cases
            failures = adv_df[(adv_df['pred'] != y_true) | (adv_df['pred_prob'] < 0.6)]
            fail_path = os.path.join(PROC, "adv_failures.parquet")
            failures.to_parquet(fail_path, index=False)
            print(f"\n Đã lưu {len(failures)} mẫu thất bại vào: {fail_path}")
        else:
            print("Không tìm thấy cột label gốc để so sánh accuracy.")

    except Exception as e:
        print(f" Lỗi trong quá trình đánh giá: {e}")
```

```python
# === Cell E: logging_monitoring ===
!pip install -q prometheus_client psutil
import time, psutil, json, os
from prometheus_client import start_http_server, Summary, Gauge

LOG_CSV = os.path.join(PROC, "inference_log.csv")
# Prometheus metrics
INF_TIME = Summary('inference_latency_seconds', 'Inference latency')
INF_GAUGE = Gauge('last_inference_latency', 'Last inference latency seconds')
MEM_GAUGE = Gauge('process_memory_mb', 'Process memory MB')

# start exporter on :8000
start_http_server(8000)
print("Prometheus exporter running on http://localhost:8000/metrics")

def log_inference(texts, model_predict_fn):
    times=[]
    for t in texts:
        st = time.time()
        _ = model_predict_fn([t])
        dt = time.time()-st
        times.append(dt)
        MEM_GAUGE.set(psutil.Process().memory_info().rss/1024/1024)
        INF_GAUGE.set(dt)
    # append to CSV
    df = pd.DataFrame({"ts":[time.time()]*len(times), "latency":times})
    if not os.path.exists(LOG_CSV):
        df.to_csv(LOG_CSV, index=False)
    else:
        df.to_csv(LOG_CSV, mode='a', header=False, index=False)
    print("Logged", len(times), "inferences ->", LOG_CSV)

# Quick test: run on some samples (use predict_texts defined above)
log_inference(adv_texts[:20], lambda x: predict_texts(x))
```
